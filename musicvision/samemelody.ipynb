```python
import os
import matplotlib.pyplot as plt
import numpy as np
from pydub import AudioSegment
import wave


sound = AudioSegment.from_mp3("./zuizhendemeng.mp3")
sound.export("zuizhen",format="wav")

sound2 = AudioSegment.from_mp3("./renshenghechubuxiangfeng.mp3")
sound2.export("rensheng",format="wav")

sound2 = AudioSegment.from_mp3("./yisibugua.mp3")
sound2.export("yisi",format="wav")



import wave
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os
from PIL import Image 
 
    
f = wave.open("./zuizhen",'rb')
params = f.getparams()
nchannels, sampwidth, framerate, nframes = params[:4]
strData = f.readframes(nframes)#
waveData = np.fromstring(strData,dtype=np.int16)#int
waveData = waveData*1.0/(max(abs(waveData)))#wave
waveData = np.reshape(waveData,[nframes,nchannels]).T
f.close()
# plot the wave
plt.figure(2)
ax1=plt.subplot(211)#21  
ax2=plt.subplot(212)#22  

plt.sca(ax1)  
plt.specgram(waveData[0],Fs = framerate, scale_by_freq = True, sides = 'default')
plt.ylabel('Frequency(Hz)')
plt.xlabel('Time(s)')
pic = plt.set_cmap('hot')


plt.sca(ax2)  
plt.specgram(waveData[0],Fs = framerate, scale_by_freq = True, sides = 'default')
plt.ylabel('Frequency(Hz)')
plt.xlabel('Time(s)')
plt.set_cmap('gray')
plt.show()

f = wave.open("./rensheng",'rb')
params = f.getparams()
nchannels, sampwidth, framerate, nframes = params[:4]
strData = f.readframes(nframes)#
waveData = np.fromstring(strData,dtype=np.int16)#int
waveData = waveData*1.0/(max(abs(waveData)))#wave
waveData = np.reshape(waveData,[nframes,nchannels]).T
f.close()
# plot the wave
plt.figure(2)
ax1=plt.subplot(211)#21  
ax2=plt.subplot(212)#22  

plt.sca(ax1)  
plt.specgram(waveData[0],Fs = framerate, scale_by_freq = True, sides = 'default')
plt.ylabel('Frequency(Hz)')
plt.xlabel('Time(s)')
pic = plt.set_cmap('hot')


plt.sca(ax2)  
plt.specgram(waveData[0],Fs = framerate, scale_by_freq = True, sides = 'default')
plt.ylabel('Frequency(Hz)')
plt.xlabel('Time(s)')
plt.set_cmap('gray')
plt.show()
```
